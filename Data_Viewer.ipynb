# Data Display
This notebook displays data found in the repository. The data displayed include:

- the particle size distributions for the three bed types (acrylic, gravel, and sand-gravel beds), 
- the suspended sediment concentration time series for each run, 
- the change in the suspended sediment particle size distribution from the beginning to the end of the depositional phase, and 
- the change in the suspended sediment distribution statistics through the depositional phase. 

The hydraulic conditions associated with each time series can be found in the “Summary Tables” directory.

Author: Kyle Strom (June 2018)

```python
# Load the needed packages
%pylab inline
%config InlineBackend.figure_format='retina' # hig-res plots for a Retina display
import pandas as pd
```

## Bed grain size distribution Data

```python
# gravel bed
ds_gr_mm = array([9.525,6.35,4.75,3.35,2.36,1.8,0.85])
perc_pass_gr = array([99.30,98.39,72.74,15.56,2.73,0.08,0.00])

# sand 
ds_sand_mm = array([1.7,1.4,1,0.71,0.5,0.3,0.25,0.18,0.15,0.075])
perc_pass_sand = array([99.25,98.08,92.21,66.23,42.64,21.41,14.34,6.71,3.73,0.64])

# sand-gravel bed 
ds_sdgr_mm = array([6.3,4.75,3.35,2.36,1.7,1.4,1,0.71,0.5,0.3,0.25,0.18,0.15,0.075])
perc_pass_sdgr = array([99.95,92.75,52.01,28.97,22.73,21.60,19.88,19.36,12.57,6.32,4.05,1.85,1.23,0.15])

# plot
fig, ax = plt.subplots(figsize=(4,3)) # size in inches
ax.plot(ds_gr_mm, perc_pass_gr, 's-', color='C1', label="Gravel")
ax.plot(ds_sand_mm, perc_pass_sand, 'o-', color='C7', label="Sand")
ax.plot(ds_sdgr_mm, perc_pass_sdgr, 'v-',color='C2', label="Sand-Gravel")
ax.set_xlabel(r'$d_{s}$ [mm]', fontsize=10)
ax.set_ylabel('Percent Passing By Weight [%]', fontsize=10)
ax.semilogx()

ax.axvline(x=2,color='k', linestyle='--', alpha=0.5)
ax.text(1.3,65, "Sand/\nGravel\nBreak", backgroundcolor='white', 
        clip_on='True', multialignment='center', alpha=0.75) #weight = 'bold', 

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
ax.legend(loc=2, fontsize=8); # more examples on how to controll the legend are given below
fig.tight_layout();
```

## Load and plot the concentration time series

```python
Cplexi = pd.read_csv('Time Series/C_mgL_TimeSeries_Acrylic.csv')
Cgravel = pd.read_csv('Time Series/C_mgL_TimeSeries_Gravel.csv')
Csndgr = pd.read_csv('Time Series/C_mgL_TimeSeries_SandGravel.csv')

depend = 580 # start of the depositional phase

fig, ax = plt.subplots(figsize=(7,3)) # size in inches
ax.plot(Cgravel, color='C1',alpha=1)
ax.plot(Csndgr, color='C2',alpha=1)
ax.plot(Cplexi,color='C0',alpha=1)

ax.set_xlim(0,1100)
ax.axvline(x=depend,color='k', linestyle='--', alpha=0.5, label='End of Dep Phase')
ax.set_xlabel(r'Time $t$ [min]', fontsize=10)
ax.set_ylabel(r'Concentration $C$ [mg/L]', fontsize=10)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax.legend(["Acrylic", "Gravel",'Sand-Gravel'],ncol=3, loc=1,fontsize=8)
leg = ax.get_legend()
leg.legendHandles[0].set_color('C0')
leg.legendHandles[1].set_color('C1')
leg.legendHandles[2].set_color('C2')

ax.text(160,-80, "Deposition Phase", weight = 'bold', alpha=0.75)
ax.text(700,-80, "Entrainment Phase", weight = 'bold', alpha=0.75)

fig.tight_layout();
# savefig(output+'C_timeseries_all.pdf',bbox_inches="tight", pad_inches=0.005);
```

## Load and plot the suspended sediment PSD statistics with time

```python
Dx_susp = pd.read_csv('Time Series/SS_PSD_Stats_TimeSeries.csv')

fig, ax = plt.subplots(figsize=(4,3.05)) # size in inches
ax.plot(Dx_susp['t_hr'][0:9]*60,Dx_susp['d90_um'][0:9], 'o',label=r'$d_{90}$')
ax.plot(Dx_susp['t_hr'][0:9]*60,Dx_susp['d84_um'][0:9], 's',label=r'$d_{84}$')
ax.plot(Dx_susp['t_hr'][0:9]*60,Dx_susp['d50_um'][0:9], 'v',label=r'$d_{50}$')
ax.plot(Dx_susp['t_hr'][0:9]*60,Dx_susp['d16_um'][0:9], '+',label=r'$d_{16}$')

ax.set_xlabel(r'Time $t$ [min]', fontsize=10)
ax.set_ylabel(r'Size $d_x$ [µm]', fontsize=10)
ax.semilogx()
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
ax.set_xlim(15,600)
ax.set_ylim(0,25)
ax.legend(loc=1, fontsize=9); # more examples on how to controll the legend are given below
ax.text(20,22,'SG:LS:HC')
fig.tight_layout();
```

## Load and plot the suspended sediment PSD over Deposition Phase

```python
PSD_susp = pd.read_csv('Time Series/SS_PSD_Start_End.csv')
# PSD_susp.head(3)

# plot and save the figure 
fig, ax = plt.subplots(figsize=(4,3)) # size in inches
ax.plot(PSD_susp['di_i'],PSD_susp['frac_i'],'o-',color='C7', label='Initial PSD at $t=t_{0}$')
ax.plot(PSD_susp['di_f'],PSD_susp['frac_f'],'s-',color='C0',label='Final PSD at $t=t_{f}$')

ax.set_xlabel(r'Size $d$ [µm]', fontsize=10)
ax.set_ylabel(r'Fraction [-]', fontsize=10)
#ax.semilogx()
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
ax.legend(loc=1, fontsize=10); # more examples on how to controll the legend are given below
ax.set_ylim(-0.015,0.18)
ax.text(3,0.1575,'SG:LS:HC')
fig.tight_layout();
```
